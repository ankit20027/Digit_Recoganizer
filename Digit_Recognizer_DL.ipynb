```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras as keras
```

```python
df_test = pd.read_csv("https://raw.githubusercontent.com/ankit20027/Digit_Recoganizer/main/test.csv")
df_train = pd.read_csv("https://raw.githubusercontent.com/ankit20027/Digit_Recoganizer/main/train.csv")
```

```python
df_train.head()
```

Train split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train.drop(columns=['label']),df_train['label'], test_size=0.2, random_state=22)
```

Main test split

```python
X_train = df_train.drop(columns=['label'])
y_train = df_train['label']
X_test = df_test
```

```python
len(X_test)
```

```python
X_train.head()
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(256, input_shape=(28*28,), activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train,y_train, verbose=1, epochs=100, batch_size=32)
```

```python
df = pd.DataFrame()
df["ImageId"] = list(range(1,len(X_test)+1))
df["Label"] = np.argmax(model.predict(X_test),axis=1)
df.to_csv("final(32).csv", header=True, index=False)
```
